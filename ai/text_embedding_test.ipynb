```python
import torch
from transformers import AutoModel, AutoTokenizer

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

phobert = AutoModel.from_pretrained("vinai/phobert-base-v2").to(device)
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base-v2")

# INPUT TEXT MUST BE ALREADY WORD-SEGMENTED!
def get_phobert_features(sentences):

    input_ids = [tokenizer.encode(sentence, padding='longest', truncation=True, return_tensors='pt') for sentence in sentences]

    print(input_ids)


    with torch.no_grad():
        features = phobert(input_ids)  # Models outputs are now tuples

    return features  # First element of the tuple: last_hidden_states
```

```python
texts = ["Tôi là sinh viên trường đại học Bách Khoa Hà Nội", "Hôm nay trời nắng đẹp quá", "Tôi yêu Việt Nam"]

features = get_phobert_features(texts)

print(features[0].shape)  # (batch_size, sequence_length, hidden_size)
```
